**In this notebook I am trying to use three alogorithms to solve a Time series problem.**
> * **xgboost:** Trying to solve the time series problem as a supervised learning problem. Break the time variable into many variables such as day, wekkday, is_weekend etc and use decision tree to predict the future.
* **Linear Regression:** Also a supervised learning technique. Uses the same training dataset as xgboost but really useful in trend capturing. You can see during implementation xgboost fails to capture the trend as the slope of trend changes with time.
* **fbprophet:** Unlike prevoius methods, this is a time-series analysis method. fbprophet is a new package open sourced by facebook team. Trying this method for first time.

Finally you ensemble the predictions from these algorithms using simple average to get the final result.

```python
import sys
from math import sqrt
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.cross_validation import KFold
from sklearn import ensemble
from sklearn import linear_model as lm
from sklearn.metrics import mean_squared_error as mse
import xgboost as xgb
```

### XGBOOST

```python
def runXGB(train_X, train_y, test_X, test_y=None):
        params = {}
        params["objective"] = "reg:linear"
        params["eta"] = 0.01
        params["min_child_weight"] = 8
        params["subsample"] = 0.9
        params["colsample_bytree"] = 0.8
        params["silent"] = 1
        params["max_depth"] = 8
        params["seed"] = 1
        plst = list(params.items())
        num_rounds = 1000

        xgtrain = xgb.DMatrix(train_X, label=train_y)
        xgtest = xgb.DMatrix(test_X)
        model = xgb.train(plst, xgtrain, num_rounds)
        pred_test_y = model.predict(xgtest)
        return pred_test_y

def rmse(act_y, pred_y):
    return np.sqrt(mse(act_y, pred_y))
```

```python
data_path = "~/"
train_file = data_path + "Train.csv"
test_file = data_path + "Test.csv"

# Reading the csv file into pandas dataframe #
train_df = pd.read_csv(train_file)
test_df = pd.read_csv(test_file)
```

```python
train_df.head()
```

```python
print("Converting to date format")
train_df["Date"] = (pd.to_datetime(train_df["Datetime"], format="%d-%m-%Y %H:%M"))
test_df["Date"] = (pd.to_datetime(test_df["Datetime"], format="%d-%m-%Y %H:%M"))
# Getting the dv and id values #
y = np.array(train_df.Count.values)
test_id = test_df.ID.values

print("Processing Date field..")
train_df["DayOfMonth"] = train_df["Date"].apply(lambda x: x.day)
test_df["DayOfMonth"] = test_df["Date"].apply(lambda x: x.day)
train_df["Hour"] = train_df["Date"].apply(lambda x: x.hour)
test_df["Hour"] = test_df["Date"].apply(lambda x: x.hour)
train_df["WeekDay"] = train_df["Date"].apply(lambda x: x.weekday())
test_df["WeekDay"] = test_df["Date"].apply(lambda x: x.weekday())
train_df["DayCount"] = train_df["Date"].apply(lambda x: x.toordinal())
test_df["DayCount"] = test_df["Date"].apply(lambda x: x.toordinal())
#train_df['is_weekend'] = train_df["WeekDay"].apply(lambda x: 1 if x==5 or x==6 else 0)
#test_df['is_weekend'] = test_df["WeekDay"].apply(lambda x: 1 if x==5 or x==6 else 0)

train = train_df.drop(["Datetime","ID","Count","Date"], axis=1)
test = test_df.drop(["Datetime","ID","Date"], axis=1)

# Dropping the columns that are not needed #
train_df.drop(["Datetime","ID","Count","Date"], axis=1, inplace=True)
test_df.drop(["Datetime","ID","Date"], axis=1, inplace=True)
```

```python
split = 5112
val_df = train_df.iloc[-split:,:]
val_y = y[-split:]
train_df = train_df.iloc[:-split,:]
train_y = y[:-split]
```

```python
train_df.shape,val_df.shape
```

```python
# Running the xgb model #
preds = runXGB(train_df, train_y, val_df)
preds = preds.astype('int')
print("Validation RMSE : {}".format(rmse(val_y,preds)))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# plot baseline and predictions
plt.plot(val_y)
plt.plot(preds)
plt.show()
```

```python
preds = runXGB(train_df, train_y, test_df)
preds = preds.astype('int')
# Saving the predictions #
sample = pd.read_csv(data_path + "Sample_Submission.csv")
sample["Count"] = preds
sample["ID"] = test_id
sample.to_csv(data_path+"sub_xgb.csv", index=False)
```

### Linear Regression model

```python
print("One hot encoding..")
temp_train_arr = np.empty([train.shape[0],0])
temp_test_arr = np.empty([test.shape[0],0])
cols_to_drop = []
for var in train.columns:
    if var in ["Hour", "WeekDay"]:
        print(var)
        lb = LabelEncoder()
        full_var_data = pd.concat((train[var],test[var]),axis=0).astype('str')
        temp = lb.fit_transform(np.array(full_var_data))
        train[var] = lb.transform(np.array( train[var] ).astype('str'))
        test[var] = lb.transform(np.array( test[var] ).astype('str'))

        cols_to_drop.append(var)
        ohe = OneHotEncoder(sparse=False)
        ohe.fit(temp.reshape(-1,1))
        temp_arr = ohe.transform(train[var].reshape(-1,1))
        temp_train_arr = np.hstack([temp_train_arr, temp_arr])
        temp_arr = ohe.transform(test[var].reshape(-1,1))
        temp_test_arr = np.hstack([temp_test_arr, temp_arr])

train = train.drop(cols_to_drop, axis=1)
test = test.drop(cols_to_drop, axis=1)
train = np.hstack( [np.array(train),temp_train_arr]).astype("float")
test = np.hstack( [np.array(test),temp_test_arr]).astype("float")
print(train.shape, test.shape)
```

```python
# Use the lastest data #
train_X = np.array(train)[12000:]
train_y = y[12000:]
test_X = np.array(test)
```

```python
# Train the linear model and predict on test data #
reg = lm.LinearRegression()
reg.fit(train_X, train_y)
preds = reg.predict(test_X).astype('int')

# writing to out file #
sample = pd.read_csv(data_path + "Sample_Submission.csv")
sample["Count"] = preds
sample["ID"] = test_id
sample.to_csv(data_path+"sub_lr.csv", index=False)
```
